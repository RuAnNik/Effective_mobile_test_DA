```python
#Задание 1: Изоморфизмы
def is_isomorphic(s, t):
    if len(s) != len(t):
        return False
    map = {}
    reverse_map = {}
    
    for char_s, char_t in zip(s, t):
        if char_s in map:
            if map[char_s] != char_t:
                return False
        else:
            map[char_s] = char_t
            if char_t in reverse_map:
                return False
            reverse_map[char_t] = char_s
    
    return True
s = 'paper' 
t = 'title' 
print(is_isomorphic(s, t))
#временная сложность алгоритма составляет O(N), где N - это длина входных строк.
#пространственная сложность алгоритма составляет O(M), где M — число уникальных символов в строке.
```

```python
#Задание 2: Натуральная последовательность
def missnig_number(nums):
    for i in range(1,len(nums)+1):
        if i !=nums[i-1]:
            return i
nums = [1, 2, 3, 4, 5, 7, 8, 9, 10, 11]
print(missnig_number(nums))
#временная сложность равна O(N), где N — длина массива.
#пространственная сложность равна O(1).
```

```python
#Задание 3: Факторизация
def prime_factors(n):
    factors = []
    d = 2
    while d**2 <= n:
        while (n % d == 0):
            factors.append(d)
            n //= d
        d += 1
    if n > 1:
        factors.append(n)
        print(factors)
n = 56
prime_factors(n)

#временная сложность равна O(n^(0.5)).
#пространственная сложность равна O(k), где k — количество простых делителей числа n.
```
